Data to Graphs

```python
import pandas as pd
import matplotlib.pyplot as plt


# ============================================================
# LOAD DATA
# ============================================================

def load_data_from_csv(file_path):
    df = pd.read_csv(file_path)
    return df


def load_sample_data():
    # Example table (you can replace this)
    data = {
        "Feature1": [10, 20, 30, 40, 50],
        "Feature2": [3, 7, 5, 9, 11],
        "Feature3": [100, 150, 120, 180, 200]
    }
    return pd.DataFrame(data)


# ============================================================
# BASIC VISUALIZATIONS
# ============================================================

def plot_line(df):
    df.plot()
    plt.title("Line Plot of Features")
    plt.xlabel("Index")
    plt.ylabel("Values")
    plt.grid(True)
    plt.show()


def plot_bar(df):
    df.plot(kind="bar")
    plt.title("Bar Chart")
    plt.xlabel("Index")
    plt.ylabel("Values")
    plt.show()


def plot_histograms(df):
    df.hist()
    plt.suptitle("Histograms of Features")
    plt.show()


def plot_scatter(df, x_col, y_col):
    plt.scatter(df[x_col], df[y_col])
    plt.title(f"Scatter Plot: {x_col} vs {y_col}")
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.show()


def plot_box(df):
    df.plot(kind="box")
    plt.title("Box Plot")
    plt.show()


def plot_correlation_heatmap(df):
    corr = df.corr()
    plt.imshow(corr)
    plt.title("Correlation Heatmap")
    plt.colorbar()
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
    plt.yticks(range(len(corr.columns)), corr.columns)
    plt.show()


# ============================================================
# MAIN
# ============================================================

if __name__ == "__main__":

    # Option 1: Load your CSV file
    # df = load_data_from_csv("your_file.csv")

    # Option 2: Use sample data
    df = load_sample_data()

    print("Data Preview:")
    print(df.head())

    # Plots
    plot_line(df)
    plot_bar(df)
    plot_histograms(df)
    plot_box(df)

    # Scatter (choose two columns)
    plot_scatter(df, "Feature1", "Feature2")

    # Correlation
    plot_correlation_heatmap(df)
```
